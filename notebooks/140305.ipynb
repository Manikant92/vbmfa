```python
import numpy as np
np.set_printoptions(precision=4, suppress=True)
import scipy.io as io
import sys
sys.path.append('../src')
%load_ext autoreload
%autoreload 2
import copy

import model
import vbmfa
```

```python
def make_model(P=3, Q=2, S=5, N=3, n_update=1, rnd_hyper=True, update_fun=None):
    np.random.seed(0)
    y = np.random.multivariate_normal(mean=np.zeros(P), cov=np.eye(P), size=N)
    y = np.transpose(y)
    h = model.Hyper(P, Q, S)
    if rnd_hyper:
        h.init_rnd()
    m = model.Model(h, y)
    for i in range(n_update):
        m.update()
    if not update_fun is None:
        update_fun(m)
    mm = vbmfa.from_model(m)
    io.savemat('../../vbmfa/data.mat', mm)
    return [m, mm]
```

```python
def print_nu(m):
    for s in range(m.S):
        print 'q_nu[{:d}]'.format(s)
        print m.q_nu[s]

def print_x(m):
    for s in range(m.S):
        print 'q_x[{:d}]'.format(s)
        print m.q_x[s]
        
def print_lm(m):
    for s in range(m.S):
        print '\nq_lm[{:d}]'.format(s)
        print m.q_lm[s]
```
