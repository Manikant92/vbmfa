```python
import numpy as np
np.set_printoptions(precision=4, suppress=True)
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import sys
import copy
sys.path.append('../src')
%load_ext autoreload
%autoreload 2

import model
import data
```

```python
def plot_fit(costs):
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(costs, linewidth=2, marker='s',markersize=5, markerfacecolor='red')
    ax.grid()
    ax.set_xlabel('Iteration')
    ax.set_ylabel('MSE')
    
def plot_y(models):
    n = len(models)
    ncols = 4
    nrows=int(np.ceil(n/float(ncols)))
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5*ncols, 5*nrows))
    ax = ax.ravel()
    for i in range(n):
        ax[i].scatter(y[0,:], y[1,:], color='blue')
        yp = models[i].predict_y()
        ax[i].scatter(yp[0,:], yp[1,:], color='red')
        ax[i].set_title('Iteration {:d}'.format(i))
        
def plot_lambda(model, s=0):
    X = model.q_lm[s].l.mean
    fig = plt.figure(figsize=[6, 4])
    ax = fig.add_axes([0.0, 0.0, 1.0, 0.75])
    ax.imshow(X, interpolation='nearest', cmap=cm.RdBu, aspect='auto')
    ax.set_yticks(range(X.shape[0]))
    ax.set_xticks(range(X.shape[1]))
    
    pre = model.q_nu[s].expectation()
    ax = fig.add_axes([0.005, 0.8, 0.995, 0.2])
    ax.bar(range(len(pre)),pre)
    ax.axis('off')
```

```python
def update(m):
    m.update_lm()
    m.update_x()
    m.update_nu()
```

```python
np.random.seed(0)
y = data.generate_clusters(P=2, Q=0, mean0=[[-10.0, -10.0], [10.0, 10.0], [-10.0, +10.0], [+10.0, -10.0]], N=[10, 10, 10, 10])
m = model.Model(model.Hyper(2, 2, 4), y)
m.init_complex(hard=True)
costs, models = m.infer(eps=0.05, maxit=7, update=update)
plot_fit(costs)
plot_y(models)
```
