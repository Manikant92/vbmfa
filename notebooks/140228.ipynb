```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import sys
sys.path.append('../src')

import model
%load_ext autoreload
%autoreload 2
```

```python
def plot_fit(costs):
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(costs, linewidth=2, marker='s',markersize=5, markerfacecolor='red')
    ax.grid()
    ax.set_xlabel('Iteration')
    ax.set_ylabel('MSE')
    
def plot_y(models):
    n = len(models)
    ncols = 4
    nrows=int(np.ceil(n/float(ncols)))
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5*ncols, 5*nrows))
    ax = ax.ravel()
    for i in range(n):
        ax[i].scatter(y[0,:], y[1,:], color='blue')
        yp = models[i].predict_y()
        ax[i].scatter(yp[0,:], yp[1,:], color='red')
        ax[i].set_title('Iteration {:d}'.format(i))
        
def plot_lambda(model, s=0):
    X = model.q_lm[s].l.mean
    fig = plt.figure(figsize=[6, 4])
    ax = fig.add_axes([0.0, 0.0, 1.0, 0.75])
    ax.imshow(X, interpolation='nearest', cmap=cm.RdBu, aspect='auto')
    ax.set_yticks(range(X.shape[0]))
    ax.set_xticks(range(X.shape[1]))
    
    pre = model.q_nu[s].expectation()
    ax = fig.add_axes([0.005, 0.8, 0.995, 0.2])
    ax.bar(range(len(pre)),pre)
    ax.axis('off')
```

```python
def print_model(model):
    print 'MSE: %f' % (model.mse())
    print 'Lambda: %s' % (np.mean(model.q_lm[0].l.mean, 0).__str__())
    print 'Mu: %s' % (model.q_lm[0].m.mean.__str__())
    print 'X: %s' % (np.mean(model.q_x[0].mean, 1).__str__())
    
def print_models(models):
    for i in range(len(models)):
        print 'Model {:d}:'.format(i)
        print_model(models[i])
        print
```

```python
np.random.seed(0)
mr = model.Model(model.Hyper(2), 10)
mr.q_lm[0].l.mean.fill(0.0)
mr.q_lm[0].m.mean.fill(10.0)
mr.q_x[0].mean.fill(0.0)
y = mr.sample_y()
m = model.Model(model.Hyper(2), y)
mses, models = m.infer(maxit=3)
plot_fit(mses)
plot_y(models)
```

```python
print_models(models)
```
